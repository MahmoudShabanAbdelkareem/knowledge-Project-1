```python
def english_to_predicate(sentence):
    words = sentence.split()

    if words[0] == "All" or words[0] == "every":
        subject = words[1]
        adjective = words[3]
        return "Ax(" + subject + "(x) -> " + adjective + "(x))"

    return "Format not supported"

def predicate_to_english(expression):
    if "Ax" in expression:
        step1 = expression.replace("Ax(", "")
        step2 = step1.replace("))", "")
        parts = step2.split("(x) -> ")

        # Added a check to prevent IndexError if the '->' pattern is missing
        if len(parts) < 2:
            return "Format not supported: Could not parse subject and adjective. Check for '->' pattern."

        subject = parts[0]
        adjective = parts[1].replace("(x", "")
        return "All " + subject + "s are " + adjective + "."

    return "Format not supported"

print("Type 'e' to convert English to Logic")
print("Type 'p' to convert Logic to English")
choice = input("Your choice: ")

if choice == 'e':
    eng_text = input("English: ")
    print("Logic: " + english_to_predicate(eng_text))
if choice == 'p':
    logic_text = input("Logic: ")
    print("English: " + predicate_to_english(logic_text))
```
